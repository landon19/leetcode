## Sequential Search 循序搜尋

```python
data=[1,2,6,3,5]
def seq_search(data,key):
    for i in data:
        if i == key:
            print('find key')
            break
        
seq_search(data,3)
```

## 二元搜尋法 Binary Search 需先排序

```python
def binary_search(data,key):
    low = 0
    upper = len(data)-1
    while low<= upper:
        mid =int( (low+upper)/2)
        if data[mid]<key:
            low = mid+1
        elif data[mid]>key:
            upper = mid -1
        else:
            return mid
            
    return print("None")

data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
binary_search(data,3)
```

## Exponential Search 指數搜尋 (用在未知長度)

```python
def exponential_search(data,key):
    low= 0
    upper=len(data)-1
    while low<= upper:
        mid = int((upper-low)*(key-data[low])/(data[upper]-data[low]))+low)
       
        if mid<low or mid>upper:
            break
        if mid < key:
            low = mid+1
        elif mid < key:
            upper = mid-1
        else:
            return mid
    return -1
    
data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exponential_search(data,3)   
```
